```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('Dates').getOrCreate()
```

```
df = spark.read.csv('/FileStore/tables/AAPL.csv',header=True,inferSchema= True)
```

```
df.head(3)
```

```
df.select(['Date','Open']).show()
```

```
from pyspark.sql.functions import (dayofmonth,hour,dayofyear,month,
                                   year,weekofyear,format_number,date_format)
```

```
df.select(dayofmonth(df['Date']),dayofyear(df['Date'])).show()
```

```
df.select(year(df['Date'])).show()
```

```
newDf=df.withColumn('Year',year(df['Date']))
```

```
newDf.head(4)
```

```
res=newDf.groupBy("Year").mean().select(["Year","avg(Close)"]).alias("AvgC")
```

```
res.show(5)
```

```
newre=res.withColumnRenamed('avg(Close)','AvgClose')
```

```
newre.select(['Year',format_number('AvgClose',2).alias('AvgClose')]).show()
```
