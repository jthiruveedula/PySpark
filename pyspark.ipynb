```
from pyspark.sql.types import (StructField,StringType,
                             IntegerType,StructType)
```

```
data_schema = [StructField('SaleNumber',StringType(),True),
StructField('DocketNumber',StringType(),True),
StructField('ID',IntegerType(),True),
StructField('AttorneyName',StringType(),True),
StructField('Plaintiff',StringType(),True),
StructField('Defendant',StringType(),True),
StructField('SaleType',StringType(),True),
StructField('SaleDate',StringType(),True),
StructField('SaleStatus',StringType(),True),
StructField('PPDate',StringType(),True),
StructField('Street',StringType(),True),
StructField('City',StringType(),True),
StructField('State',StringType(),True),
StructField('ZIPCode',StringType(),True),
StructField('Address',StringType(),True),
StructField('Municipality',StringType(),True),
StructField('CostsTaxes',StringType(),True),
StructField('ServiceCompleted',StringType(),True),
StructField('Delivered3129',StringType(),True),
StructField('ReadyForSale',StringType(),True),
StructField('latitude',StringType(),True),
StructField('longitude',StringType(),True),
StructField('AdditionalNotes',StringType(),True),
StructField('RecordDate',StringType(),True)]
```

```
df=spark.read.format("csv").option('Header',True).option('sep',',').option('schema',data_schema).load("/FileStore/tables/sample.csv")
```
