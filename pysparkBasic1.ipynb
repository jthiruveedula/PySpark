```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName("ops").getOrCreate()
```

```
df = spark.read.csv("/FileStore/tables/AAPL.csv",inferSchema=True,header=True)
```

```
df.show(5)
df.printSchema()
```

```
df.head(5)[0]
```

```
df.filter('Close < 500').select(['Open','High']).show()
```

```
df.filter(df['Close']>70).select(['Open','High','Close']).show()
```

```
df.filter((df['Open'] > 65) & (df['Close']<70 )).show()
```

```
#~ is indicating "not"
df.filter((df['Open'] > 65) & ~(df['Close']<70 )).show()
```

```
#using collect in real world we will have many columns so better using collect
cl=df.filter((df['Open'] > 65) & (df['Close'] =='70.004997' )).collect()
```

```
clR=cl[0]
```

```
clR.asDict()
clR.asDict()['Volume']
```
